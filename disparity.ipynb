```python
for metric in ['auc', 'recall', 'precision', 'accuracy']:
    for attack_type in attack_types:
        if 'default' in subgroup_disparity_dicts_by_attack_type[metric][attack_type]:
            subgroup_disparity_dicts_by_attack_type[metric][attack_type]['MAR'] = subgroup_disparity_dicts_by_attack_type[metric][attack_type]['default']
            del subgroup_disparity_dicts_by_attack_type[metric][attack_type]['default']
```

```python
def plot_all_attack_disparity_on_one_sensitive_attribute(subgroup_disparity_dicts_by_attack_type, sensitive_column='MAR', subgroup_column='RAC1P', metric='auc'):
    all_attack_mar_rac1p_dict = {}
    for attack_type in attack_types:
        all_attack_mar_rac1p_dict[attack_type] = subgroup_disparity_dicts_by_attack_type[attack_type][sensitive_column][subgroup_column]

    all_attack_mar_rac1p_df = pd.DataFrame(all_attack_mar_rac1p_dict).T

    all_attack_mar_rac1p_df.plot(kind='bar', figsize=(10, 5), title=f'{metric} disparity on {sensitive_column} by attack type (Subgroup {subgroup_column})')
```

# Disparity of all attack on RAC1P

```python
all_sensitive_columns = ['MAR', 'DEAR', 'DEYE', 'DREM', 'DPHY']
for sensitive_column in all_sensitive_columns:
    plot_all_attack_disparity_on_one_sensitive_attribute(subgroup_disparity_dicts_by_attack_type['auc'], sensitive_column=sensitive_column, subgroup_column='RAC1P', metric='auc')
```
